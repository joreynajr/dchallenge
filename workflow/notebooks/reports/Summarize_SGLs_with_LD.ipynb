```python
import os 
import glob
import pandas as pd 
os.chdir('/mnt/BioHome/jreyna/jreyna/projects/dchallenge/')
```

## Load Original data

```python
globs = 'results/main/sgls/*/*/*/*/master.tsv'
globs = glob.glob(globs)
```

```python
coloc_data = []
for fn in globs:
    
    file_info = fn.split('/')  
    
    
    gwas_source = file_info[3]
    eqtl_source = file_info[4]
    ge_source = file_info[5]
    loop_source = file_info[6]
           
    num_lines = len(open(fn).readlines())
    if num_lines > 0: 
        
        #print(fn)
        
        tdf = pd.read_table(fn)
        tdf.loc[:, 'gwas_source'] = gwas_source
        tdf.loc[:, 'eqtl_source'] = eqtl_source
        tdf.loc[:, 'ge_source'] = ge_source
        tdf.loc[:, 'loop_source'] = loop_source

        coloc_data.append(tdf)
```

```python
coloc_data = pd.concat(coloc_data)
coloc_data.loc[:, 'sgl_id'] = coloc_data['sid'] + '-' + coloc_data['geneid']
coloc_summary = coloc_data.groupby(['gwas_source', 'eqtl_source', 'ge_source', 'loop_source'])#.pos.count()
```

```python
coloc_data.head()
```

```python
coloc_summary = coloc_summary.sgl_id.count().to_frame()
coloc_summary.columns = ['Number of SGLS in Coloc']
coloc_summary
```

## Load SGL-LD data

```python
globs = 'results/main/GRCh37/sgls/ldpairs/*/*/*/*/script_version/master.tsv'
globs = glob.glob(globs)
```

```python
ld_data = []
for fn in globs:
    
    file_info = fn.split('/')  
    
    
    gwas_source = file_info[5]
    eqtl_source = file_info[6]
    ge_source = file_info[7]
    loop_source = file_info[8]
       
    num_lines = len(open(fn).readlines())
    if num_lines > 0: 
        
        #print(fn)
        
        tdf = pd.read_table(fn)
        tdf.loc[:, 'gwas_source'] = gwas_source
        tdf.loc[:, 'eqtl_source'] = eqtl_source
        tdf.loc[:, 'ge_source'] = ge_source
        tdf.loc[:, 'loop_source'] = loop_source

        ld_data.append(tdf)
```

```python
ld_data = pd.concat(ld_data)
ld_data.loc[:, 'sgl_id'] = ld_data['sid'] + '-' + ld_data['geneid']
ld_summary = ld_data.groupby(['gwas_source', 'eqtl_source', 'ge_source', 'loop_source'])#.pos.count()
```

```python
ld_summary = ld_summary.sgl_id.count().to_frame()
ld_summary.columns = ['Number of SGLs in ld']
ld_summary
```
